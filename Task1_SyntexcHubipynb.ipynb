```python
import numpy as np
import pandas as pd
import time
```

```python
np.random.seed(42)

students = 1000

math_scores = np.random.randint(40, 100, students)
science_scores = np.random.randint(35, 100, students)
english_scores = np.random.randint(45, 100, students)
attendance = np.random.randint(60, 100, students)

dataset = np.column_stack((math_scores, science_scores, english_scores, attendance))

print("Dataset Shape:", dataset.shape)
print("\nFirst 5 rows:\n", dataset[:5])
```

```python
# Access first student record
print("First Student Record:", dataset[0])

# Access Math scores (first column)
math_column = dataset[:, 0]
print("First 10 Math Scores:", math_column[:10])

# Slice first 5 students
print("First 5 Students:\n", dataset[:5])

# Students with index 100 to 110
print("Students 100-110:\n", dataset[100:111])
```

```python
# Add 5 bonus marks to all subjects (except attendance)
bonus_dataset = dataset.copy()
bonus_dataset[:, :3] += 5

print("After Adding Bonus:\n", bonus_dataset[:5])
```

```python
# Average score per student
avg_per_student = np.mean(dataset[:, :3], axis=1)

# Average score per subject
avg_per_subject = np.mean(dataset[:, :3], axis=0)

print("Average per Student (first 5):", avg_per_student[:5])
print("Average per Subject (Math, Science, English):", avg_per_subject)
```

```python
print("Mean:", np.mean(dataset[:, :3]))
print("Median:", np.median(dataset[:, :3]))
print("Standard Deviation:", np.std(dataset[:, :3]))
print("Maximum Score:", np.max(dataset[:, :3]))
print("Minimum Score:", np.min(dataset[:, :3]))
```

```python
reshaped = dataset[:, :3].reshape(1000, 3)
print("Reshaped Shape:", reshaped.shape)
```
